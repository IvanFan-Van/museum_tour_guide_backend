```python
from pathlib import Path

doc_dir = Path(os.getcwd()).parent.resolve() / "data" / "objectifying_china" / "docs"
```

```python
# get all children file paths
docs = list(doc_dir.rglob("*.*"))
```

```python
docs[0].stem
```

```python
import json


def check(id):
    with open(docs[id], "r", encoding="utf-8") as f:
        content = json.load(f)
    print(f"===== File name: {docs[id].stem} =====\n\n")
    print(content["documents"])


def check_iter():
    for i in range(len(docs)):
        yield check(i)


ch_iter = check_iter()
next(ch_iter)
```
